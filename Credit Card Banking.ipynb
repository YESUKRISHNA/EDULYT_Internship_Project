## Identifying where the repayment is more than the spend to give them a credit of 2% of their surplus amount in the next month billing. 

```python
#now lets import spend and repayment sheet of excel
spend = pd.read_excel(r"C:\Users\PAIDI\Desktop\Project 1 Edulyt\Credit Banking\Spend.xlsx")
repayment = pd.read_excel(r"C:\Users\PAIDI\Desktop\Project 1 Edulyt\Credit Banking\Repayment.xlsx")
```

```python
spend
```

```python
spend.drop("Sl No:", axis=1,inplace=True)
```

```python
spend.rename(columns = {'Costomer':'Customer'},inplace=True)
```

```python
spend
```

```python
repayment
```

```python
repayment.drop("SL No:",axis=1,inplace=True)
```

```python
repayment.rename(columns = {'Costomer':'Customer'},inplace=True)
```

```python
repayment
```

```python
surplus_data = spend.merge(repayment, on=["Customer","Month"],how="outer",suffixes=("_spend","_repayment"))
```

```python
surplus_data
```

```python
surplus_data["Surplus"]=surplus_data["Amount_repayment"]-surplus_data["Amount_spend"]
```

```python
surplus_data
```

```python
credit=surplus_data.loc[surplus_data["Surplus"]>0,"Surplus"]*0.02
```

```python
credit
```

```python
surplus_data.loc[surplus_data["Surplus"]>0,"Credit"]=credit
```

```python
surplus_data
```

```python
surplus_data.drop("Surplus",axis=1,inplace=True)
```

```python
surplus_data
```

# 5) Which age group is spending more money? 

```python
merged_data1["Age"].max()
```

```python
age_bins=[0,18,25,30,35,40,45,50,55,60,65,70,75,79]
age_labels=["<18","18-25","25-30","30-35","35-40","40-45","45-50","50-55","55-60","60-65","65-70","70-75","75-79"]
```

```python
merged_data1["Age Group"] = pd.cut(merged_data1["Age"], bins = age_bins, labels = age_labels)
```

```python
age_group_spend=merged_data1.groupby("Age Group")["Amount"].sum()
age_group_spend
```

```python
max_age_group=age_group_spend.idxmax()
```

# 8) Monthly profit for the bank.

```python
surplus_data["yy-mm"] = surplus_data["Month"].dt.strftime("%y-%m")
```

# Correction Here
### surplus_data["Amount_repayment"] - surplus_data["Amount_spend"] - surplus_data["Credit"].fillna(0)

```python
surplus_data["Profit"] = surplus_data["Amount_repayment"] - surplus_data["Amount_spend"] + surplus_data["Credit"].fillna(0)
```

```python
surplus_data.head(4)
```

```python
bank_monthly_profit = surplus_data.groupby("yy-mm")["Profit"].sum()
```

```python
bank_monthly_profit
```

```python
bank_monthly_profit[bank_monthly_profit>0].reset_index()
```
